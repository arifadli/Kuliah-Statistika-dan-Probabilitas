```python
import numpy as np

# 1. Menyiapkan data berdasarkan tabel di gambar (image_b67a87.png)
# Variabel Independen (X1, X2) dan Dependen (Y)
x1 = np.array([45, 42, 44, 45, 43, 46, 44, 45, 44, 43])
x2 = np.array([16, 14, 15, 13, 13, 14, 16, 16, 15, 15])
y  = np.array([29, 24, 27, 25, 26, 28, 30, 28, 28, 27])

n = len(y)

# 2. Menghitung komponen untuk Persamaan Normal
sum_y = np.sum(y)
sum_x1 = np.sum(x1)
sum_x2 = np.sum(x2)
sum_x1y = np.sum(x1 * y)
sum_x2y = np.sum(x2 * y)
sum_x1_sq = np.sum(x1**2)
sum_x2_sq = np.sum(x2**2)
sum_x1x2 = np.sum(x1 * x2)

print("--- Hasil Perhitungan Komponen ---")
print(f"n: {n}, ΣY: {sum_y}, ΣX1: {sum_x1}, ΣX2: {sum_x2}")
print(f"ΣX1Y: {sum_x1y}, ΣX2Y: {sum_x2y}")
print(f"ΣX1²: {sum_x1_sq}, ΣX2²: {sum_x2_sq}, ΣX1X2: {sum_x1x2}\n")

# 3. Menyusun Matriks Persamaan (Metode Persamaan Normal)
# Matriks A berisi koefisien dari a, b1, dan b2
# Matriks B berisi hasil di ruas kiri (ΣY, ΣX1Y, ΣX2Y)
A = np.array([
    [n,      sum_x1,    sum_x2],
    [sum_x1, sum_x1_sq, sum_x1x2],
    [sum_x2, sum_x1x2,  sum_x2_sq]
])

B = np.array([sum_y, sum_x1y, sum_x2y])

# 4. Menyelesaikan sistem persamaan linear (Mencari nilai a, b1, b2)
solusi = np.linalg.solve(A, B)
a, b1, b2 = solusi

print("--- Hasil Estimasi Parameter ---")
print(f"Nilai a  (Konstanta) : {a:.4f}")
print(f"Nilai b1 (Koefisien) : {b1:.4f}")
print(f"Nilai b2 (Koefisien) : {b2:.4f}")

# 5. Persamaan Regresi
print(f"\nPersamaan Regresi: Y = {a:.3f} + {b1:.3f}X1 + {b2:.3f}X2")
```

```python
import pandas as pd
import numpy as np

# Membuat dataframe agar terlihat seperti tabel di gambar
data = {
    'X1': [45, 42, 44, 45, 43, 46, 44, 45, 44, 43],
    'X2': [16, 14, 15, 13, 13, 14, 16, 16, 15, 15],
    'Y':  [29, 24, 27, 25, 26, 28, 30, 28, 28, 27]
}
df = pd.DataFrame(data)

# Menghitung kolom tambahan
df['X1Y'] = df['X1'] * df['Y']
df['X2Y'] = df['X2'] * df['Y']
df['X1X2'] = df['X1'] * df['X2']
df['X1_sq'] = df['X1']**2
df['X2_sq'] = df['X2']**2

# Membuat Baris Total (Summary)
summary = df.sum()
print("--- TABEL RINGKASAN STATISTIK ---")
print(df)
print("\n--- TOTAL (SUM) ---")
print(summary)

# Selesaikan Persamaan Normal menggunakan Numpy (Matrix Inverse)
A = np.array([
    [len(df), summary['X1'], summary['X2']],
    [summary['X1'], summary['X1_sq'], summary['X1X2']],
    [summary['X2'], summary['X1X2'], summary['X2_sq']]
])
B = np.array([summary['Y'], summary['X1Y'], summary['X2Y']])

a, b1, b2 = np.linalg.solve(A, B)

print(f"\nHasil Akhir: a = {a:.3f}, b1 = {b1:.3f}, b2 = {b2:.3f}")
```

```python
import numpy as np

# 1. SETUP DATA (Sesuai Tabel)
n = 10
sum_y = 272
sum_x1 = 441
sum_x2 = 147
sum_x1y = 12005
sum_x2y = 4013
sum_x1_sq = 19461
sum_x2_sq = 2173
sum_x1x2 = 6485

print("=== LANGKAH 1: MENYUSUN PERSAMAAN NORMAL ===")
print(f"(1) {sum_y} = {n}a + {sum_x1}b1 + {sum_x2}b2")
print(f"(2) {sum_x1y} = {sum_x1}a + {sum_x1_sq}b1 + {sum_x1x2}b2")
print(f"(3) {sum_x2y} = {sum_x2}a + {sum_x1x2}b1 + {sum_x2_sq}b2")
print("-" * 50)

# 2. ELIMINASI a DARI PERSAMAAN (1) & (2) -> Menjadi Persamaan (4)
# Kita kalikan Pers (1) dengan (sum_x1 / n) agar koefisien 'a' sama
faktor12 = sum_x1 / n
p1_scaled_12 = sum_y * faktor12
p1_a_12 = n * faktor12
p1_b1_12 = sum_x1 * faktor12
p1_b2_12 = sum_x2 * faktor12

# Pers (2) - Pers (1)_scaled
res_y_4 = sum_x1y - p1_scaled_12
res_b1_4 = sum_x1_sq - p1_b1_12
res_b2_4 = sum_x1x2 - p1_b2_12

print("=== LANGKAH 2: ELIMINASI 'a' (Pers 1 & 2) ===")
print(f"Pers (1) dikali {faktor12}: {p1_scaled_12:.2f} = {p1_a_12:.2f}a + {p1_b1_12:.2f}b1 + {p1_b2_12:.2f}b2")
print(f"Pers (2)           : {sum_x1y:.2f} = {sum_x1:.2f}a + {sum_x1_sq:.2f}b1 + {sum_x1x2:.2f}b2")
print(f"HASIL (Pers 4)     : {res_y_4:.2f} = {res_b1_4:.2f}b1 + {res_b2_4:.2f}b2")
print("-" * 50)

# 3. ELIMINASI a DARI PERSAMAAN (1) & (3) -> Menjadi Persamaan (5)
faktor13 = sum_x2 / n
p1_scaled_13 = sum_y * faktor13
p1_a_13 = n * faktor13
p1_b1_13 = sum_x1 * faktor13
p1_b2_13 = sum_x2 * faktor13

# Pers (3) - Pers (1)_scaled
res_y_5 = sum_x2y - p1_scaled_13
res_b1_5 = sum_x1x2 - p1_b1_13
res_b2_5 = sum_x2_sq - p1_b2_13

print("=== LANGKAH 3: ELIMINASI 'a' (Pers 1 & 3) ===")
print(f"Pers (1) dikali {faktor13}: {p1_scaled_13:.2f} = {p1_a_13:.2f}a + {p1_b1_13:.2f}b1 + {p1_b2_13:.2f}b2")
print(f"Pers (3)           : {sum_x2y:.2f} = {sum_x2:.2f}a + {sum_x1x2:.2f}b1 + {sum_x2_sq:.2f}b2")
print(f"HASIL (Pers 5)     : {res_y_5:.2f} = {res_b1_5:.2f}b1 + {res_b2_5:.2f}b2")
print("-" * 50)

# 4. ELIMINASI b1 DARI PERSAMAAN (4) & (5) -> MENCARI b2
# Kalikan Pers (4) dengan (res_b1_5 / res_b1_4)
faktor45 = res_b1_5 / res_b1_4
p4_scaled_y = res_y_4 * faktor45
p4_scaled_b1 = res_b1_4 * faktor45
p4_scaled_b2 = res_b2_4 * faktor45

# Pers (5) - Pers (4)_scaled
final_y = res_y_5 - p4_scaled_y
final_b2_coeff = res_b2_5 - p4_scaled_b2

b2 = final_y / final_b2_coeff

print("=== LANGKAH 4: ELIMINASI 'b1' (Pers 4 & 5) UNTUK b2 ===")
print(f"Pers (4) dikali {faktor45:.4f}: {p4_scaled_y:.2f} = {p4_scaled_b1:.2f}b1 + {p4_scaled_b2:.2f}b2")
print(f"Pers (5)               : {res_y_5:.2f} = {res_b1_5:.2f}b1 + {res_b2_5:.2f}b2")
print(f"HASIL AKHIR b2: {final_y:.4f} / {final_b2_coeff:.4f}")
print(f"Maka, b2 = {b2:.4f}")
print("-" * 50)

# 5. SUBSTITUSI BALIK UNTUK b1 (Menggunakan Pers 4)
# b1 = (y_4 - (coeff_b2_4 * b2)) / coeff_b1_4
b1 = (res_y_4 - (res_b2_4 * b2)) / res_b1_4

print("=== LANGKAH 5: SUBSTITUSI b2 KE PERS (4) UNTUK b1 ===")
print(f"{res_y_4:.2f} = {res_b1_4:.2f}b1 + {res_b2_4:.2f}({b2:.4f})")
print(f"b1 = ({res_y_4:.2f} - {res_b2_4 * b2:.2f}) / {res_b1_4:.2f}")
print(f"Maka, b1 = {b1:.4f}")
print("-" * 50)

# 6. SUBSTITUSI BALIK UNTUK a (Menggunakan Pers 1)
# a = (sum_y - (sum_x1 * b1) - (sum_x2 * b2)) / n
a = (sum_y - (sum_x1 * b1) - (sum_x2 * b2)) / n

print("=== LANGKAH 6: SUBSTITUSI b1 & b2 KE PERS (1) UNTUK a ===")
print(f"{sum_y} = 10a + {sum_x1}({b1:.4f}) + {sum_x2}({b2:.4f})")
print(f"a = ({sum_y} - {sum_x1*b1:.2f} - {sum_x2*b2:.2f}) / 10")
print(f"Maka, a = {a:.4f}")
print("-" * 50)

print("\nKESIMPULAN:")
print(f"Persamaan: Y = {a:.3f} + {b1:.3f}X1 + {b2:.3f}X2")
```
